## Шаг 2. Построим отчеты и подсчитаем метрики

### 2.1.*Продукт (таблица visits)*

 ### 1) *Сколько людей пользуются в день, неделю, месяц?*

 Чтобы расчитать метрики **DAU**(количество уникальных пользоватлей в день), **WAU**(количество уникальных пользоватлей в неделю) и **MAU**(количество уникальных пользоватлей в месяц), нам нужно создать в таблице **visits** новые столбцы(*'session_date'*, *'session_week'*, *'session_month'* и *'session_year'*), в которые мы выделим  полную дату, неделю, месяц и год, соответственно.

```python
visits['session_date'] = visits['start_ts'].dt.date
visits['session_week'] = visits['start_ts'].dt.week
visits['session_month'] = visits['start_ts'].dt.month
visits['session_year'] = visits['start_ts'].dt.year
visits.head()
```

Перейдем к расчетy самих метрик:

```python
dau_total = visits.groupby('session_date').agg({'uid': 'nunique'}).mean()
dau_total
```

```python
wau_total = visits.groupby(['session_year', 'session_week']).agg({'uid': 'nunique'}).mean()
wau_total
```

```python
mau_total = visits.groupby(['session_year', 'session_month']).agg({'uid': 'nunique'}).mean()
mau_total
```

Во всех трех случаях мы видим довольно хорошие показатели, при том,что речь идет не о социальных сетях, а о сервисе покупки билетов, куда люди заходят отнюдь не каждый день.

Теперь визуализируем пользовательскую активность по всему данному нам временному периоду, но распределим ее по устройствам, с которого смотрели сайт, а также по рекламным источникам, с которого пользователи переходили на него.

```python
visits['activity_month'] = visits['start_ts'].astype('datetime64[M]')
```

```python
visits_by_device=visits.pivot_table(
        index='activity_month', 
        columns='device', 
        values='uid', 
        aggfunc='nunique')

visits_by_device.plot(grid=True, figsize = (20,9))
plt.title('Пользовательская активность по месяцам(в зависимости от устройства)', fontsize=20)
plt.gca().set(xlabel='месяц', ylabel='количество сессий')
```

По построенному нами графику мы видим, что пик активности приходится на конец 2017ого года. Можно сделать предположение, что пользователи покупали билеты на новогодние мероприятия (как детские, так и взрослые). Также по всей видимости,сайтом удобнее пользоваться с компьютера, т.к. линия *'desktop'* на графике выше, чем *'touch'* (мобильные устройства, планшеты). Возможно, есть основания поработать над улучшением мобильной версии сайта.

```python
visits_by_source=visits.pivot_table(
        index='activity_month', 
        columns='source_id', 
        values='uid', 
        aggfunc='nunique')

visits_by_source.plot(grid=True, figsize = (20,9))
plt.title('Пользовательская активность по месяцам(в зависимости от рекламного источника)', fontsize=20)
plt.gca().set(xlabel='месяц', ylabel='количество сессий')
```

Так же, как и на графике с устройствами, мы видим,что пик активности приходится на конец 2017ого года,причем это касается всех источников,кроме №7 и №10. У этих источников самые низкие показатели(у источника №7 показатели и вовсе на 0) - их необходимо пересмотреть. А вот источники №3 и №4 лидируют, причем время от времени, пересекают друг друга и меняются местами - то есть можно сказать, что оба источника работают хорошо.

### 2) *Сколько сессий в день?*

Чтобы найти эту метрику, нам необходимо разделить число сессий на количество пользователей за период:

```python
daily_user_sessions = visits.groupby('session_date').agg({'uid': ['count','nunique']})
                      
daily_user_sessions.columns = ['n_sessions', 'n_users']

daily_user_sessions['sessions_per_user'] = daily_user_sessions['n_sessions'] / daily_user_sessions['n_users']
```

```python
daily_user_sessions['sessions_per_user'].mean()
```

В среднем пользователь поводит на сайте в день одну сессию. Для сайта с тематикой, подобной нашей (продажа билетов на культурные мероприятия) это хороший показатель.

### 3) *Сколько длится одна сессия?*

Найдём **ASL**(cредняя продолжительность сессии) — посчитаем длину всех сессий и расчитаем **моду**(значение, встречающееся в выборке чаще всего):

```python
visits['session_duration_sec'] = (visits['end_ts'] - visits['start_ts']).dt.seconds
visits.head()
```

```python
asl = visits['session_duration_sec'].mode()
asl
```

Продолжительность совсем небольшая. Возможно, все дело в недоработанной мобильной версии - сессии со смартфонов или планшетов могут длиться не более минуты. 

Построим гистограмму, чтобы подробнее рассмотреть этот вопрос:

```python
visits_asl_device = visits.pivot_table(index='activity_month',columns='device',values='session_duration_sec')
visits_asl_device.plot(kind='bar', grid=True, label=['desktop','touch'],figsize=(14, 8))

plt.legend()
plt.title('Продолжительность сессии(в зависимости от устройства)')
plt.gca().set(xlabel='месяц', ylabel='секунды')
plt.show()
```

По графику мы видим, что пользователи, зашедшие на сайт с компьютера, проводят там около 10 минут - этого времени достаточно, чтобы выбрать место и купить билет, если пользователь целенаправленно заходит на страницу мероприятия, на которое расчитывает пойти. Показатель немного повышается в ноябре - возможно, это связано с большим разнообразием новогодних мероприятий.

Также по этой гистограмме можно убедиться в выводе, который мы сделали выше - нужно усовершенствовать мобильную версию сайта. С компьютеров совершается в разы больше сессий, чем с мобильных телефонов\планшетов. 

### 4) *Как часто люди возвращаются?*

Определим событие и период — на их основе сформируем когорту. Возьмём дату, когда пользователь впервые проявил активность на сайте. Эти данные запишем в объект Series **first_activity_date**. Объединим его с таблицей **visits**:

```python
first_activity_date = visits.groupby(['uid'])['start_ts'].min()
first_activity_date.name = 'first_activity_date'
visits = visits.join(first_activity_date,on='uid') 
visits.head()
```

Будем формировать когорты по данным за неделю. Получим день начала недели, за которую произошло событие. Он станет идентификатором недели. Параметр *unit* метода **pd.to_timedelta** задаёт единицу измерения — в нашем случае дня. Вычтем из даты порядковый номер дня: 

```python

visits['first_activity_month'] = visits['first_activity_date'].astype('datetime64[M]')
visits.head()
```

Теперь для каждой строки датафрейма можно рассчитать **lifetime** пользователя в рамках когорты. В качестве делителя обратимся к константе **np.timedelta64(1,'W')**: 

```python
visits['cohort_lifetime'] = visits['activity_month'] - visits['first_activity_month']
visits['cohort_lifetime'] = visits['cohort_lifetime'] / np.timedelta64(1,'M')
visits['cohort_lifetime'] = visits['cohort_lifetime'].round().astype('int')
visits.head()
```

Сгруппируем данные по когорте и lifetime. Посчитаем для каждой когорты количество активных пользователей на определённую «неделю жизни»:

```python
cohorts = visits.groupby(['first_activity_month','cohort_lifetime']).agg({'uid':'nunique'}).reset_index() 
```

Чтобы найти ***Retention Rate***, нужно сперва получить число пользователей, изначально бывших в когорте, и на него разделить число пользователей в каждую следующую неделю.Найдём исходное количество пользователей в когорте. Возьмём их число на нулевую неделю:

```python
initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_activity_month','uid']]
initial_users_count
```

Переименуем столбец *'user_id'* в *'cohort_users'*:

```python
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'}) 
initial_users_count
```

Объединим данные по когортам с исходным количеством пользователей в когорте:

```python
cohorts = cohorts.merge(initial_users_count,on='first_activity_month') 
cohorts
```

Рассчитаем ***Retention Rate***. Разделим количество активных пользователей в каждую из недель на исходное число пользователей в когорте:

```python
cohorts['retention'] = cohorts['uid']/cohorts['cohort_users'] 
cohorts
```

Построим сводную таблицу и создадим тепловую карту:

```python
retention_pivot = cohorts.pivot_table(index='first_activity_month',columns='cohort_lifetime',values='retention',aggfunc='sum') 
retention_pivot.info()
```

```python
retention_pivot.index = retention_pivot.index.strftime("%m/%Y")
sns.set(style='white')
plt.figure(figsize=(15, 10))
plt.title('Retention Rate(по когортам)')
sns.heatmap(retention_pivot, annot=True, fmt='.2f', vmax=0.15, linewidths=1, linecolor='grey')
plt.gca().set(xlabel='lifetime', ylabel='первый месяц активности')
plt.show()
```

По тепловой таблице мы видим,что ***Ratention Rate*** убывает с течением времени - за год показатель упал с 1 до 0.04. С каждой новой когортой поведение пользователей меняется. Этому может быть несколько причин:

1) рекламные источнки работают не очень хорошо, и пользователи стали покупать билеты на мероприятия напрямую;

### 4) *Сколько денег приносят? (LTV)*

Поскольку эту метрику нам нужно будет посчитать в разбивке по рекламным источникам, нам нужно добавить новый столбец *'first_activity_source'* c данными об источнике, с которого пользователь первый раз пришел на сайт:

```python
first_activity_source = visits.sort_values(by = 'start_ts').groupby('uid')['source_id'].first()
first_activity_source.name = 'first_activity_source'
visits = visits.merge(first_activity_source,on='uid', how = 'left') 
visits.head()
```

Создадим небольшую таблицу(**visits_short**) с нужными нам столбцами из таблицы **visits**:

```python
visits_short = visits[['uid','first_activity_source']]
visits_short
```

Для того, чтобы при объединении нам получить таблицу с исходным количеством строк, мы избавимся от дубликатов в таблице **visits_short**:

```python
visits_short.duplicated().sum()
```

```python
visits_short_drop = visits_short.drop_duplicates()
```

Проверяем:

```python
visits_short_drop.duplicated().sum()
```

```python
print(f'Контрольная ячейка с размером таблицы orders: {orders.shape}')
```

Методом объединения (***merge***) свяжем две таблицы (**orders** и **visits_short_drop**) в одну (**new_orders**):

```python
new_orders = pd.merge(orders,visits_short_drop, on='uid', how='left')
new_orders
```

```python
print(f'Контрольная ячейка с размером таблицы orders: {new_orders.shape}')
```

<a class="anchor" id="21-bullet">

```python
visits_short['uid'].value_counts()
```

В этой таблице собраны все возможные заходы пользователя, а не 1. Если нужно найти источник первого захода пользователя, то можно искать первый по времени, то есть таблицу визитов отсортировать по дате и при группировке по посетителю можно брать первый источник.

Посчитаем количество покупателей за каждый месяц и на каждый рекламный источник:

```python
cohort_sizes = new_orders.groupby(['first_activity_source','first_purchase_month']).agg({'uid': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_activity_source','first_purchase_month', 'n_buyers']
cohort_sizes
```

```python
cohorts = new_orders.groupby(['first_activity_source','first_purchase_month','purchase_month']).agg({'revenue': 'sum'}).reset_index()
cohorts
```

Объединим две таблицы - **cohort_sizes** и  **cohorts** -  и назовем общую таблицу **report**:

```python
report = pd.merge(cohort_sizes, cohorts, on = ['first_activity_source','first_purchase_month'])
report
```

Добавим столбец *'age'* с информацией о "возрасте" когорты - для этого найдем разницу между месяцем покупки и первым месяцем покупки: 

```python
report['age'] = (report['purchase_month'] - report['first_purchase_month']) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')                
```

Перейдем к расчету **LTV**.  

**LTV** считают не из выручки, а из валовой прибыли. Для того,чтобы вычислить валовую прибыль, нам нужно произведение маржинальности и выручки. В данном случае у нас нет конкретной информации о маржинальности, поэтому примем ее за 1. 

Соответственно, чтобы найти **LTV**, нам нужно выручку разделить на количество пользователей:

```python
report['ltv'] = report['revenue'] / report['n_buyers']
report
```

С помощью цикла построим тепловые таблицы в разбивке на рекламные источники:

```python
for source_id in report.first_activity_source.unique():
 
    output = report.query('first_activity_source == @source_id').pivot_table(
        index= 'first_purchase_month', 
        columns='age', 
        values='ltv', 
        aggfunc='mean').round().cumsum(axis=1)
    
    output.index = output.index.strftime("%m/%Y")
 
    plt.figure(figsize=(16, 6))
    ax = sns.heatmap(output, linewidths=0.3, annot=True, fmt='.2f', linecolor='gray', cmap="mako_r")
    ax.set_title('LTV по рекламному источнику '+str(source_id))
    plt.xlabel('lifetime')
    plt.ylabel('месяц первой покупки(когорты)')
    plt.show()
```

Меньше всего принес денег источник **№7** - это сходится с показателями графика, отображающего пользовательскую активность. С источника **№10** "пропадают" покупатели, и цифры этого источника совсем невысоки. Возможно, следует пересмотреть работу с ними.

Больше всего принесли источники **№1, №2, №3** и **№4**. И рост показателей небольшой, но стабильный - от месяцу к месяцу.

### 2) *Сколько стоило привлечение одного покупателя из каждого источника?*

Для этого нам нужно объединить данные из двух таблиц. Сгруппируем данные из таблицы **costs** по рекламным источникам(*'source_id'*) и месяцу (*'dt_activity'*), затраченные на рекламу деньги суммируем. Назовем сгруппированную таблицу **costs_grouped_source**. 

```python
costs_grouped_source = costs.groupby(['source_id','dt_activity']).agg({'costs':'sum'}).reset_index()
costs_grouped_source
```

Переименуем столбец *'dt_activity'*, чтобы мы смогли объединить таблицы:

```python
report['cac'] = report['costs'] / report['n_buyers']
report
```

```python
report.groupby('first_activity_source')['cac'].mean() \
            .nlargest(7)
```

### 3)*На сколько окупились расходы? (ROI)*

Теперь у нас есть данные для расчета **ROI**.

```python
report['roi'] = report['ltv'] / report['cac']
report
```

```python
output_roi = report.pivot_table(
        index= 'first_purchase_month', 
        columns='age', 
        values='roi', 
        aggfunc='mean').round(2).cumsum(axis=1)
output_roi.fillna('')
```

```python
output_roi.index = output_roi.index.strftime("%m/%Y")
 
plt.figure(figsize=(16, 6))
ax = sns.heatmap(output_roi, linewidths=0.3, annot=True, fmt='.2f', linecolor='gray', cmap="mako_r")
ax.set_title('ROI')
plt.xlabel('lifetime')
plt.ylabel('месяц первой покупки(когорты)')
plt.show()
```

По тепловой карте мы видим, что расходы окупились во всех когортах, кроме январской. 10 дней новогодних прадников дают о себе знать:

1) на все крупные мероприятия билеты куплены заранее, возможны мелкие покупки (например, билеты в кино);

2) новогодние праздники дают о себе знать в финансовом плане - много расходов + десятидневные каникулы неблагоприятно сказываются на бюджете, многим приходится в январе урезать траты, и первое, на чем люди будут экономить - это, конечно, на вылазках в театры\кино\музеи.

Источнику **№3**  следует, по моему мнению, пересмотреть алгоритм - окупаемость близка к 0. "Афише" же, в силу специфики направления, нужно привлекать как можно больше пользователей.

На графике, где отображена пользовательская активность в зависимости от рекламных источников, пик активности приходится на конец 2017ого года,причем это касается всех источников,кроме №7 и №10. * Сколько сессий в день?

В среднем пользователь поводит на сайте в день одну сессию. Для сайта с тематикой, подобной нашей (продажа билетов на культурные мероприятия) это хороший показатель.

* Сколько длится одна сессия? (**ASL**)

Мы нашли **моду** столбца *'session_duration_sec'* (продолжительность сессии в секнудах), она составляет **60 секунд**. Продолжительность совсем небольшая. Возможно, все дело в мобильной версии - сессии со смартфонов или планшетов могут длиться не более минуты, а сейчас люди все-таки чаще пользуются смартфонами, нежели компьютерами. Чтобы убедиться в этом, мы построили гистограмму продолжительности сессии в зависимости от устройств.

По этой гистограмме можно убедиться в выводе, который мы сделали выше - нужно усовершенствовать мобильную версию сайта. С компьютеров совершается в разы больше сессий, чем с мобильных телефонов\планшетов. Больше всего сессий имеет продолжительность от 0 до 1000 секунд(16 минут) - этого времени хватает как раз на выбор места и покупку билета на мероприятие, когда люди уже знают, куда хотят пойти (гораздо меньше пользователей поводят на сайте больше 16 минут(полчаса и более) - скорее всего,это те, кто выбирает, на какой спектакль\фильм\концерт\выставку им пойти).

* Как часто люди возвращаются? (**Ratention Rate**)

По тепловой таблице мы видим,что ***Ratention Rate*** убывает с течением времени - за год показатель упал с 1 до 0.04. С каждой новой когортой поведение пользователей меняется. Этому может быть несколько причин:

1) рекламные источнки работают не очень хорошо, и пользователи стали покупать билеты на мероприятия напрямую;

2) сейчас люди очень часто используют смартфоны, а как мы уже выяснили на предыдущих этапах нашего исследования - над мобильной версией "Афиши" следует еще поработать.

### 3.2.Продажи

* Когда люди начинают покупать?

Мы столбец *'purchase_start'*, в котором посчитали разницу между датой первой покупки и датой первого посещения. Оба расчета показывают, что самое повторяющееся значение этого столбца - **0**. То есть чаще всего пользователи совершают покупку в тот же день, когда первый раз заходят на сайт. Это хороший признак:

* видимо, сайт удобен в эксплуатации(по крайней мере с компьютера);

* покупатели сразу находят то, что ищут;

* хорошо работают рекламные источнки (например, были куплены верные слова, по которым пользователь сразу находит нужное ему событие в поисковке и с его помощью переходит на "Афишу").

* Сколько раз покупают за период?

По построенной нами тепловой таблице можно увидеть,что в среднем на одного покупателя приходится по 1 покупке в месяц. Учитывая то, что "Афиша" занимается продажей билетов на мероприятия - это нормальный показатель. Многие могут позволить выбраться куда-либо 1, максимум  2 раза в месяц (например, в силу нехватки свободного времени или ограниченного бюджета). Покупки можно увеличивать за счет прироста новых покупателей. 

* Какой средний чек?

Самый большой средний чек приходится на декабрь 2017 года - время покупки билетов на новогодние мероприятия. С января показатели опять падают.

* Сколько денег приносят? (**LTV**)

Мы посчитали метрику **LTV** с помощью объединения таблиц. Построили сводную таблицу **output**, где сгруппировали  значения столбца  *'ltv'* по первому рекламному источнику, с которого зашел покупатель, и по первому месяцу активности, а затем в цикле для каждого рекламного источника построили тепловую карту.

Меньше всего принес денег источник **№7** - это сходится с показателями графика, отображающего пользовательскую активность. С источника **№10** "пропадают" покупатели, и цифры этого источника совсем невысоки. Возможно, следует пересмотреть работу с ними.

Больше всего принесли источники **№1, №2, №3** и **№4**. И рост показателей небольшой, но стабильный - от месяцу к месяцу.

К тому же летом театры, например, завершают сезон (обычно в июне), это также сокращает количество проводимых мероприятий. Нет большого смысла вкладывать большие материальные ресурсы в эти месяцы, лучше приберечь их на осенне-зимний период.

* На сколько окупились расходы? (**ROI**)

Мы расчитали **ROI**, создадали сводную таблицу, где сгруппировали среднее **ROI** по когортам(первому месяцу покупки), затем построили по ней(таблице) тепловую карту. По ней видно,что окупаемость растет от месяца к месяцу, но медленно - у "зимних" когорт она может несколько месяцев "жизни"  оставаться на одном показателе. 

По графику видно,что окупились почти все представленные нам для исследования рекламные источники. Источнику **№3**  следует, по моему мнению, пересмотреть алгоритм - окупаемость близка к 0. "Афише" же, в силу специфики направления, нужно привлекать как можно больше пользователей.
