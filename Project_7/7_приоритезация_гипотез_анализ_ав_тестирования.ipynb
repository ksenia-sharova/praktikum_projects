Вместе с отделом маркетинга мы подготовили список гипотез для увеличения выручки. Нам необходимо приоритизировать гипотезы, запустить A/B-тест и проанализировать результаты. 

## Шаг 1. Приоритизация гипотез

В начале работы импортируем все нужные нам библиотеки. Затем создадим таблицу **df_hyp**, куда поместим 9 гипотез по увеличению выручки интернет-магазина с указанными параметрами: 
* *Reach* (охват пользователей)
* *Impact* (влияние на пользователей)
* *Confidence* (насколько мы уверены в оценках других параметров)
* *Effort* (насколько сложно проверить гипотезу)

```python
import scipy.stats as stats
import pandas as pd
pd.set_option('display.max_colwidth', -1)
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
```

```python
df_hyp = pd.read_csv('/home/kseniya/Downloads/hypothesis.csv')
df_hyp
```

Приведем все названия столбцов таблицы **df_hyp** к строчному регистру:

```python
df_hyp.columns = df_hyp.columns.str.lower()
df_hyp
```

Применим фреймворк ***ICE*** (метод приоритизации, который позволяет оценить задачи более детально) для приоритизации гипотез. Мы помножим столбцы *'impact'* и *'confidence'*, а их произведение разделим на значения столбца *'efforts'*:

```python
df_hyp['ICE'] = (df_hyp['impact'] * df_hyp['confidence']) / df_hyp['efforts']
```

Создадим небольшую таблицу **df_ice**, в которой будут необходимые нам столбцы *'hypothesis'* и *'ICE'*, значения которого мы отсортируем по убаванию приоритета: 

```python
df_ice = df_hyp[['hypothesis', 'ICE']].sort_values(by='ICE', ascending=False)
df_ice
```

По резутатам данной приоритизации наиболее перспективными являются гипотезы **8**, **0** и **7**.

Теперь для сравнения применим метод **RICE** - при применении этого метода нам понадобиться столбец *'reach'*:

```python
df_hyp['RICE'] = (df_hyp['reach'] * df_hyp['impact'] * df_hyp['confidence']) / df_hyp['efforts']
```

Для оценки резутатов будем действовать по тому же принципу, что и с методом **ICE**: создадим небольшую таблицу **df_rice**, в которой будут необходимые нам столбцы *'hypothesis'* и *'RICE'*, значения которого мы отсортируем по убаванию приоритета: 

```python
df_rice = df_hyp[['hypothesis', 'RICE']].sort_values(by='RICE', ascending=False)
df_rice
```

### Вывод

Перед нами стояла задача приоритизировать гипотезы. Для этого мы применили два метода (**ICE** и **RICE**) к данным таблицы **df_hyp**.

По резутатам приоритизации методом **ICE** наиболее перспективными являются гипотезы **8**, **0** и **7**.

Как только мы добавили в расчеты охват пользователей, вперед вырвалась гипотеза **7** (причем с большим отрывом). Также в случае с методом **RICE** в тройку лидеров вошла гипотеза **2**, которой не было в ряду перспективных, когда не брали в расчет охват пользователей. Большое количество пользователей затронет гипотеза **0** (она была в числе перспективных и по результатам применения метода **ICE**). Те же результаты у гипотезы **6** - она находится на четвертом месте и по результатам применения и того, и другого методов. Соответственно,эту гипотезу тоже можно считать перспективной.

Чтобы построить графики, нужно собрать кумулятивные данные. Объявим датафрейм **cumulativeData** со столбцами:

*   ***'date'*** — дата;
*   ***'group'*** — группа A/B-теста (A или B);
*   ***'orders'*** — количество заказов в указанную дату в указанной группе;
*   ***'buyers'*** — количество пользователей, совершивших хотя бы 1 заказ в указанную дату в указанной группе;
*   ***'revenue'*** — выручка в указанную дату в указанной группе (средний чек);
*   ***'visitors'*** — количество посетителей интернет-магазина в указанную дату в определённой группе.

Создадим массив уникальных пар значений дат и групп теста методом **drop_duplicates()**:

```python
datesGroups = orders[['date','group']].drop_duplicates() 
datesGroups
```

С помощью метода **np.logical()** получим строки таблицы **orders**, дата которых меньше или равна дате элемента из **datesGroups**, а группа теста равна группе из **datesGroups**. Агрегируем значения: 
* вычислим максимальную дату; 
* для группы тоже рассчитаем максимум (хотя она будет определённой, сделаем это, чтобы столбец *'group'* не пропал из итогового вывода);
* найдём число уникальных ID заказов и пользователей;
* подсчитаем сумму средних чеков.

Применим методы к каждой строке датафрейма (**apply(axis=1)**) и отсортируем результаты по столбцам *'date'* и *'group'*:

```python
ordersAggregated = datesGroups.apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])]\
.agg({'date' : 'max', 'group' : 'max', 'transactionId' : pd.Series.nunique, 'visitorId' : pd.Series.nunique, 'revenue' : 'sum'}), axis=1) \
.sort_values(by=['date','group']) 
```

Аналогично получим агрегированные кумулятивные по дням данные о посетителях интернет-магазина:

```python
visitorsAggregated = datesGroups.apply(lambda x: visitors[np.logical_and(visitors['date'] <= x['date'], visitors['group'] == x['group'])] \
                    .agg({'date' : 'max', 'group' : 'max', 'visitors' : 'sum'}), axis=1) \
                    .sort_values(by=['date','group']) 
```

Объединим кумулятивные данные в одной таблице(**cumulativeData**) и переименуем столбцы:

```python
cumulativeData = ordersAggregated.merge(visitorsAggregated, left_on=['date', 'group'], right_on=['date', 'group'])
cumulativeData.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']
cumulativeData
```

Для построения графика разделим нашу таблицу **cumulativeData** на два датафрейма с кумулятивным количеством заказов и кумулятивной выручкой по дням и по группам А (**cumulativeRevenueA**) и В(**cumulativeRevenueB**):

```python
cumulativeRevenueA = cumulativeData[cumulativeData['group']=='A'][['date','revenue', 'orders']]
cumulativeRevenueA.head()
```

```python
cumulativeRevenueB = cumulativeData[cumulativeData['group']=='B'][['date','revenue', 'orders']]
cumulativeRevenueB.head()
```

Наконец мы можем построить график:

```python
plt.subplots(figsize=(16, 9))
plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue'], label='A')
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue'], label='B')
plt.legend() 
plt.title('График кумулятивной выручки по группам', fontsize=20)
plt.gca().set(xlabel='дата', ylabel='выручка')
```

**Вывод**: по графику видим, что выручка обеих групп увеличивается. Однако, выручка группы А увеличивается более равномерно, без резких подъемов. У выручка же группы В мы можем наблюдать резкий подъем во временном промежутке от 17 августа 2019 года до 21 августа 2019 года - это может сигнализировать о всплесках числа заказов, либо о появлении очень дорогих заказов в выборке. 

Также по этому графику мы видим, что выручка группы В больше, чем выручка группы А.

### 2.3. Построим график кумулятивного среднего чека по группам

Построим графики среднего чека по группам — разделим кумулятивную выручку на кумулятивное число заказов:

```python
plt.subplots(figsize=(16, 9))
plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue']/cumulativeRevenueA['orders'], label='A')
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue']/cumulativeRevenueB['orders'], label='B')
plt.legend()
plt.title('График кумулятивного среднего чека по группам', fontsize=20)
plt.gca().set(xlabel='дата', ylabel='средний чек')
```

**Вывод**: на данном графике мы видим рост среднего чека группы В (причем в уже известном нам по предыдущему графику временном промежутке от 17 августа 2019 года до 21 августа 2019 года мы наблюдаем резкий подъем - гопять же, можем предположить, что имел место быть крупный заказ), а к концу месяца происходит небольшой спад.  

У линии группы А мы видим снижение, затем рост - 8 августа группа А даже опережает группу В, однако затем происходит небольшой спад и  стабилизация. Но средний чек группы А все же меньше чека группы В.

### 2.4. Построим график относительного изменения кумулятивного среднего чека группы B к группе A.

Соберем из двух датафреймов (**cumulativeRevenueA** и **cumulativeRevenueB**) в один - **mergedCumulativeRevenue**:

```python
mergedCumulativeRevenue = cumulativeRevenueA.merge(cumulativeRevenueB, left_on='date', right_on='date', how='left', suffixes=['A', 'B'])
mergedCumulativeRevenue.head()
```

Строим отношение средних чеков. Добавим горизонтальную ось методом **axhline()**:

```python
plt.subplots(figsize=(16, 9))
plt.plot(mergedCumulativeRevenue['date'], (mergedCumulativeRevenue['revenueB']/mergedCumulativeRevenue['ordersB'])/(mergedCumulativeRevenue['revenueA']/mergedCumulativeRevenue['ordersA'])-1)
plt.axhline(y=0, color='black', linestyle='--') 
plt.title('График относительного изменения кумулятивного среднего чека группы B к группе A', fontsize=20)
plt.gca().set(xlabel='дата', ylabel='средний чек')
```

**Вывод**: мы видим несколько резких скачков , особенно в уже знакомом нам промежутке от 17 августа 2019 года до 21 августа 2019 года. Можно предположить,что все-таки имеет место крупный заказ или выброс, так как мы видим резкие скачки в одном и том же временном промежутке на всех трех построенных нами графиках.

## Шаг 3. Расчет статистических значимостей

### 3.1. Посчитаем статистическую значимость различий в конверсии между группами по «сырым» данным (без удаления аномальных пользователей).

Начнем с подготовки данных.

В таблицу **visitorsADaily** поместим данные о количестве пользователей, посетивших сайт за день (из группы А):

```python
visitorsADaily = visitors[visitors['group']=='A'][['date', 'visitors']]
visitorsADaily.columns = ['date', 'visitorsPerDateA']
visitorsADaily.head()
```

В таблице **visitorsACummulative** посчитаем количество пользователей до выбранной даты включительно в группе A:

```python
visitorsACummulative = visitorsADaily.apply(lambda x: visitorsADaily[visitorsADaily['date'] <= x['date']].agg({'date' : 'max', 'visitorsPerDateA' : 'sum'}), axis=1)
visitorsACummulative.columns = ['date', 'visitorsCummulativeA']
visitorsACummulative.head()
```

Те же самые действия совершаем по отношению к группе В:

```python
visitorsBDaily = visitors[visitors['group']=='B'][['date', 'visitors']]
visitorsBDaily.columns = ['date', 'visitorsPerDateB']
visitorsBDaily.head()
```

```python
visitorsBCummulative = visitorsBDaily.apply(lambda x: visitorsBDaily[visitorsBDaily['date'] <= x['date']].agg({'date' : 'max', 'visitorsPerDateB' : 'sum'}), axis=1)
visitorsBCummulative.columns = ['date', 'visitorsCummulativeB']
visitorsBCummulative.head()
```

Из таблицы **orders** нам нужно отфильтровать данные по группам (сначала А, потом В), затем получить информацию о количестве заказов в день и о ежедневной выручке:

```python
ordersADaily = orders[orders['group']=='A'][['date', 'transactionId', 'visitorId', 'revenue']]\
    .groupby('date', as_index=False)\
    .agg({'transactionId' : pd.Series.nunique, 'revenue' : 'sum'})
ordersADaily.columns = ['date', 'ordersPerDateA', 'revenuePerDateA']
ordersADaily.head()
```

В таблице **ordersACummulative** найдем суммарную выручку и суммарное количество заказов до выбранной даты (максимальное значение столбца *'date'*, то есть конец месяца) включетельно, результат отсортируем по возрастанию даты (с помощью метода **sort_values**):

```python
ordersACummulative = ordersADaily.apply(
    lambda x: \
        ordersADaily[ordersADaily['date'] <= x['date']]\
            .agg({
                'date' : 'max',
                'ordersPerDateA' : 'sum',
                'revenuePerDateA' : 'sum'}), axis=1) \
    .sort_values(by=['date'])
ordersACummulative.columns = ['date', 'ordersCummulativeA', 'revenueCummulativeA']
ordersACummulative.head()
```

Подобные таблицы создаем для группы В:

```python
ordersBDaily = orders[orders['group']=='B'][['date', 'transactionId', 'visitorId', 'revenue']]\
    .groupby('date', as_index=False)\
    .agg({'transactionId' : pd.Series.nunique, 'revenue' : 'sum'})
ordersBDaily.columns = ['date', 'ordersPerDateB', 'revenuePerDateB']
ordersBDaily.head()
```

```python
ordersBCummulative = ordersBDaily.apply(
    lambda x: \
        ordersBDaily[ordersBDaily['date'] <= x['date']]\
            .agg({
                'date' : 'max',
                'ordersPerDateB' : 'sum',
                'revenuePerDateB' : 'sum'}), axis=1) \
        .sort_values(by=['date'])
ordersBCummulative.columns = ['date', 'ordersCummulativeB', 'revenueCummulativeB']
ordersBCummulative.head()
```

Объединим все расчеты в одну таблицу - **df_merge** - по столбцу *'date'* с помощью метода **merge**:

```python
df_merge = ordersADaily.merge(ordersBDaily, left_on='date', right_on='date', how='left')\
    .merge(ordersACummulative, left_on='date', right_on='date', how='left')\
    .merge(ordersBCummulative, left_on='date', right_on='date', how='left')\
    .merge(visitorsADaily, left_on='date', right_on='date', how='left')\
    .merge(visitorsBDaily, left_on='date', right_on='date', how='left')\
    .merge(visitorsACummulative, left_on='date', right_on='date', how='left')\
    .merge(visitorsBCummulative, left_on='date', right_on='date', how='left')

df_merge.head()
```

```python
df_merge.shape
```

Мы собрали все нужные нам данные в одной таблице ( она состоит из 13 столбцов, ее длина - 31 строка). Еще раз, для наглядности, перечислим данные, содержащиеся в таблице **df_merge**:

*    *'date'* — дата;
*    *'ordersPerDateA'* — количество заказов в выбранную дату в группе A;
*    *'revenuePerDateA'* — суммарная выручка в выбранную дату в группе A;
*    *'ordersPerDateB'* — количество заказов в выбранную дату в группе B;
*    *'revenuePerDateB'* — суммарная выручка в выбранную дату в группе B;
*    *'ordersCummulativeA'* — суммарное число заказов до выбранной даты включительно в группе A;
*    *'revenueCummulativeA'* — суммарная выручка до выбранной даты включительно в группе A;
*    *'ordersCummulativeB'* — суммарное количество заказов до выбранной даты включительно в группе B;
*    *'revenueCummulativeB'* — суммарная выручка до выбранной даты включительно в группе B;
*    *'visitorsPerDateA'* — количество пользователей в выбранную дату в группе A;
*    *'visitorsPerDateB'* — количество пользователей в выбранную дату в группе B;
*    *'visitorsCummulativeA'* — количество пользователей до выбранной даты включительно в группе A;
*    *'visitorsCummulativeB'* — количество пользователей до выбранной даты включительно в группе B.

Подготовим выборки к проверке *критерием Манна-Уитни*, т.к. этот критерий устойчив к выбросам.

Создадим переменные **ordersByUsersA** и **ordersByUsersB** со столбцами *'userId'*, *'orders'*. В них для пользователей, которые заказывали хотя бы 1 раз, укажем число совершённых заказов:

```python
ordersByUsersA = orders[orders['group']=='A'].groupby('visitorId', as_index=False).agg({'transactionId' : pd.Series.nunique})
ordersByUsersA.columns = ['userId', 'orders']
ordersByUsersA.head()
```

```python
ordersByUsersB = orders[orders['group']=='B'].groupby('visitorId', as_index=False).agg({'transactionId' : pd.Series.nunique})
ordersByUsersB.columns = ['userId', 'orders']
ordersByUsersB.head()
```

Объявим переменные **sampleA** и **sampleB**, в которых пользователям из разных групп будет соответствовать количество заказов. Тем, кто ничего не заказал, будут соответствовать нули. Эти переменные должны состоять из двух частей:

1) Список с количеством заказов для каждого из пользователей: **ordersByUsersA['orders']**.

2) Нули для пользователей, которые ничего не заказывали. Их количество равно разнице между суммой посетителей и количеством записей о заказах: **df_merge['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])**.

Последовательности объединим с помощью функции **pd.concat()**:

```python
sampleA = pd.concat([ordersByUsersA['orders'],pd.Series(0, index=np.arange(df_merge['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])), \
                                                        name='orders')],axis=0)

sampleA
```

```python
sampleB = pd.concat([ordersByUsersB['orders'],pd.Series(0, index=np.arange(df_merge['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])), \
                                                        name='orders')],axis=0) 
sampleB
```

Теперь можно применить критерий Манна-Уитни.

Сформулируем гипотезы:

*   **Нулевая гипотеза**: Конверсии групп A и B равны с точки зрения средних значений.
*   **Альтернативна гипотеза**: Конверсии групп A и B различаются с точки зрения средних значений.

Чтобы получить конверсию в группе, поделим сумму заказов на число пользователей (число элементов **sampleA** и **sampleB** — это количество пользователей, сумма всех элементов — количество заказов) — найдём среднее в выборке **sampleA** и **sampleB** методом mean().Выведем относительный прирост конверсии группы B: **конверсия группы B / конверсия группы A - 1**. Округлим до трёх знаков после запятой (*p-value* также округлим до пяти знаков после запятой):

```python
print("{0:.5f}".format(stats.mannwhitneyu(sampleA, sampleB)[1]))
```

```python
print("{0:.3f}".format(sampleB.mean()/sampleA.mean()-1)) 
```

**Вывод**: видим, что  *p-value* = 0.00840, что меньше 0.05. Соответственно, мы можем отвергнуть нулевую гипотезу и сделать предположение, что статистически значимые различия в конверсии между группами А и В есть. Группа В опережает группу А примерно на **13,8%**. Делаем вывод,что ***по "сырым" данным различия между конверсиями группы А и группы В есть***.

### 3.2. Посчитаем статистическую значимость различий в среднем чеке заказа между группами по «сырым» данным.

Сформулируем гипотезы:

*   **Нулевая гипотеза**:  Статистически значимых различий в среднем чеке заказа между группами А и В нет.
*   **Альтернативная гипотеза**: Статистически значимые различия в среднем чеке заказа между группами А и В есть.

Чтобы рассчитать статистическую значимость различий в среднем чеке между сегментами, передадим критерию **mannwhitneyu()** данные о выручке с заказов, а вторым расчетом найдём относительные различия в среднем чеке между группами (эти показатели мы будем считать с помощью изначальной таблицы **orders**):

```python
print("{0:.3f}".format(stats.mannwhitneyu(orders[orders['group']=='A']['revenue'], orders[orders['group']=='B']['revenue'])[1]))
```

```python
print("{0:.3f}".format(orders[orders['group']=='B']['revenue'].mean()/orders[orders['group']=='A']['revenue'].mean()-1)) 
```

**Вывод**:  в данном вопросе мы не можем сделать однозначный вывод, так как  *p-value* = 0.365, это значительно больше 0.05, соответственно, мы можем принять нулевую гипотезу, однако расчеты разницы в среднем чеке между группами показывают, что средний чек группы В больше почти на **26%**.  

### 3.3. Посчитаем статистическую значимость различий в конверсии между группами по «очищенным» данным.

Теперь нам необходимо отфильтровать данные.

Ниже в нашем исследовании мы решили, что за крайнюю границу мы берем *2 заказа* на пользователя и заказы , не дороже *58 233.2 рублей*.

Сделаем срезы пользователей с числом заказов больше 4 — **usersWithManyOrders** и пользователей, совершивших заказы дороже 58 233.2 рублей — **usersWithExpensiveOrders**. Объединим их в таблице **abnormalUsers**:

```python
usersWithManyOrders = pd.concat([ordersByUsersA[ordersByUsersA['orders'] > 2]['userId'],\
                                 ordersByUsersB[ordersByUsersB['orders'] > 2]['userId']], axis = 0)
usersWithManyOrders
```

```python
usersWithExpensiveOrders = orders[orders['revenue'] > 58233.2]['visitorId']
usersWithExpensiveOrders
```

```python
abnormalUsers = pd.concat([usersWithManyOrders, usersWithExpensiveOrders], axis = 0).drop_duplicates().sort_values()
abnormalUsers
```

```python
abnormalUsers.shape
```

31 аномальный пользователь. Узнаем, как их действия повлияли на результаты. 

Подготовим выборки количества заказов по пользователям по группам теста:

```python
sampleAFiltered = pd.concat([ordersByUsersA[np.logical_not(ordersByUsersA['userId'].isin(abnormalUsers))]['orders'],\
                             pd.Series(0, index=np.arange(df_merge['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])),name='orders')],axis=0)
sampleAFiltered
```

```python
sampleBFiltered = pd.concat([ordersByUsersB[np.logical_not(ordersByUsersB['userId'].isin(abnormalUsers))]['orders'],\
                             pd.Series(0, index=np.arange(df_merge['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])),name='orders')],axis=0)
sampleBFiltered
```

Применим статистический критерий Манна-Уитни к полученным выборкам:

```python
print("{0:.5f}".format(stats.mannwhitneyu(sampleAFiltered, sampleBFiltered)[1]))
```

```python
print("{0:.3f}".format(sampleBFiltered.mean()/sampleAFiltered.mean()-1)) 
```

**Вывод**: *p-value* = 0.00504, это так же меньше 0.05, как и в случае с "сырыми" данными, а группа В опережает группу А даже чуть больше - на **17,4%**. Вывод можем сделать тот же, что и расчете по "сырым" данным: ***различия между конверсиями группы А и группы В есть***.

### 3.4. Посчитаем статистическую значимость различий в среднем чеке заказа между группами по «очищенным» данным.

В данном случае нам опять пригодится изначальная таблица **orders**:

```python
print("{0:.3f}".format(stats.mannwhitneyu(
    orders[np.logical_and(
        orders['group']=='A',
        np.logical_not(orders['visitorId'].isin(abnormalUsers)))]['revenue'],
    orders[np.logical_and(
        orders['group']=='B',
        np.logical_not(orders['visitorId'].isin(abnormalUsers)))]['revenue'])[1]))
```

```python
print("{0:.3f}".format(
    orders[np.logical_and(orders['group']=='B',np.logical_not(orders['visitorId'].isin(abnormalUsers)))]['revenue'].mean()/
    orders[np.logical_and(
        orders['group']=='A',
        np.logical_not(orders['visitorId'].isin(abnormalUsers)))]['revenue'].mean() - 1)) 
```

**Вывод**: *p-value* = 0.399, то есть этот показатель практически не изменился, и мы можем принять нулевую гипотезу *("Статистически значимых различий в среднем чеке заказа между группами А и В нет")*, а вот разница в средних чеках между группами существенно сократилась: теперь группа В даже несколько уступает группе А - на **2%**. Здесь и видна разница между "сырыми" и "очищенными" данными - по всей видимости, покупки, превышающие ту границу, которую мы обозначили (58233.2 рубля), совершали именно покупатели из группы В.

По этому графику видим, что конверсия группы В набирает обороты, она в целом стабильна и выше конверсии группы А.

**График относительного изменения кумулятивного среднего чека группы B к группе A**:

```python
plt.subplots(figsize=(14, 7))
plt.plot(mergedCumulativeRevenue['date'], (mergedCumulativeRevenue['revenueB']/mergedCumulativeRevenue['ordersB'])/(mergedCumulativeRevenue['revenueA']/mergedCumulativeRevenue['ordersA'])-1)
plt.axhline(y=0, color='black', linestyle='--') 
plt.title('График относительного изменения кумулятивного среднего чека группы B к группе A')
plt.gca().set(xlabel='дата', ylabel='средний чек')
```

Этот график менее стабилен (теперь уже можно предположить,что это последствия пользователей, совершивших дорогие покупки). Однако все равно средний чек группы В по этому графику выше, чем средний чек группы А. Хотя к концу месяца мы можем наблюдать небольшой спад (2%, которые мы расчитали, когда "очистили" данные).  

По результатам остальных графиков группа В также опережает группу А.

Мы применили ***критерий Манна-Уитни***, чтобы посчитаем статистическую значимость различий в конверсии и среднем чеке между группами по "сырым" и "очищенным" данным. Результаты таковы:

*  статистическая значимость различий в конверсии между группой А и группой В есть, причем и в "сырых", и в "очищенных" данных;

*  в среднем чеке до фильтрации были неоднозначные результаты (*p-value* значительно больше 0.05, соответственно, мы могли бы принять нулевую гипотезу, однако расчеты разницы в среднем чеке между группами показывали, что средний чек группы В больше почти на **26%**), а после фильтрации расчеты показали, что статистической значимости различий в среднем чеке между группой А и группой В нет.

Исходя из обнаруженных фактов и судя по произведенным нами расчетам, тест следует остановить и признать успешным. По всем графикам группа В лучше группы А, а график конверсии говорит о фиксации результата. Статистически группа В также лучше группы А.
