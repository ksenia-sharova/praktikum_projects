## Шаг 2.Считаем и добавляем данные

Нам необходимо посчитать для каждого пользователя количество: 

1. израсходованных минут по месяцам
2. израсходованного интернет-трафика
3. отправленных сообщений

Для этого мы будем использовать метод **groupby**.

### 2.1. Подготовим данные

Для начала в каждой из нужных нам таблиц (**users**, **internet**, **messages**, **calls**) создадим столбец *'month'*, в котором выделим номер месяца, когда абонент совершал звонки, пользовался интернетом и пр. 

```python
users['month'] = users['reg_date'].dt.month
users.head()
```

```python
calls['month'] = calls['call_date'].dt.month
calls.head()
```

```python
internet['month'] = internet['session_date'].dt.month
internet.head()
```

```python
messages['month'] = messages['message_date'].dt.month
messages.head()
```

Создадим переменную ***group***, по которой мы будем группировать наши таблицы для нужных нам расчетов:

```python
group = ['user_id', 'month']
```

Теперь произведем необходимые нам расчеты по месяцам.

### 2.2. Количество израсходованных минут (в месяц)

```python
calls_grouped = calls.groupby(group).agg({'duration':'sum'})
calls_grouped['duration'] = np.ceil(calls_grouped['duration'])
#calls_grouped['duration'] = calls_grouped['duration'].astype(int)
calls_grouped.head()
```

### 2.3. Количество израсходованного интернет-трафика (в месяц)

```python
internet_grouped = internet.groupby(group).agg({'mb_used':'sum'})
internet_grouped['mb_used'] = np.ceil(internet_grouped['mb_used'])
#internet_grouped['mb_used'] = internet_grouped['mb_used'].astype(int)
internet_grouped.head()
```

### 2.4. Количество отправленных сообщений (в месяц)

```python
messages_grouped = messages.groupby(group).agg({'id': 'count'}).rename(columns = {'id':'messages'})
messages_grouped.head()
```

### 2.5. Объединяем таблицы

Теперь нам понадобиться соединить несколько таблиц в одну - **grouped_df**(она будет содержать все нужные нам для исследования данные). Сделаем это поэтапно. 

* *Объединим в одну все расчитанные нами ниже помесячные таблицы по столбцу *user_id**

```python
grouped_df = calls_grouped.merge(messages_grouped, on=group, how='outer').merge(internet_grouped, on=group, how='outer')
grouped_df.head()
```

* *Присоединим таблицу  'users'*

```python
grouped_df = grouped_df.fillna(0).reset_index().merge(users, on='user_id', how='left')
grouped_df.head()
```

* *Соединяем существующую таблицу **grouped_df** с **tariffs** - это понадобится нам для расчета выручки по тарифам*

```python
grouped_df = grouped_df.merge(tariffs, left_on='tariff', right_on='tariff_name', how='left')
grouped_df.head()
```

```python
f'Количество пользователей: {grouped_df.user_id.nunique()}'
```

### 2.6. Разбиваем по тарифам

Теперь мы разобьем большую таблицу **grouped_df** на две таблицы, отфильтрованные по тарифам(**grouped_df_ultra** и **grouped_df_smart**)

```python
grouped_df_ultra = grouped_df.query('tariff == "ultra"').reset_index(drop=True)
grouped_df_ultra.head() 
```

```python
grouped_df_smart = grouped_df.query('tariff == "smart"').reset_index(drop=True)
grouped_df_smart.head() 
```

### 2.7. Расчитаем помесячную выручку с каждого пользователя

Чтобы получить помесячную выручку с каждого пользователя, сначала вычтем бесплатный лимит из суммарного количества звонков, сообщений и интернет-трафика:

```python
grouped_df_ultra['over_minutes'] = grouped_df_ultra['duration'] - grouped_df_ultra['minutes_included']
grouped_df_ultra['over_mb'] = grouped_df_ultra['mb_used'] - grouped_df_ultra['mb_per_month_included']
grouped_df_ultra['over_messages'] = grouped_df_ultra['messages'] - grouped_df_ultra['messages_included']
grouped_df_ultra.head()
```

```python
grouped_df_smart['over_minutes'] = grouped_df_smart['duration'] - grouped_df_smart['minutes_included']
grouped_df_smart['over_mb'] = grouped_df_smart['mb_used'] - grouped_df_smart['mb_per_month_included']
grouped_df_smart['over_messages'] = grouped_df_smart['messages'] - grouped_df_smart['messages_included']
grouped_df_smart.head()
```

Все полученные отрицательные значения заменим 0:

```python
grouped_df_ultra.loc[grouped_df_ultra['over_minutes'] < 0, 'over_minutes'] = 0
grouped_df_ultra.loc[grouped_df_ultra['over_mb'] < 0, 'over_mb'] = 0
grouped_df_ultra.loc[grouped_df_ultra['over_messages'] < 0, 'over_messages'] = 0
```

```python
grouped_df_smart.loc[grouped_df_smart['over_minutes'] < 0, 'over_minutes'] = 0
grouped_df_smart.loc[grouped_df_smart['over_mb'] < 0, 'over_mb'] = 0
grouped_df_smart.loc[grouped_df_smart['over_messages'] < 0, 'over_messages'] = 0
```

И,наконец, получим помесячную выручку с каждого пользователя(создадим в двух таблицах отдельный столбец *'fee'*:

```python
grouped_df_ultra['fee'] = grouped_df_ultra['rub_monthly_fee'] +\
                        grouped_df_ultra['over_minutes'] * grouped_df_ultra['rub_per_minute'] +\
                        grouped_df_ultra['over_mb'] / 1000 * grouped_df_ultra['rub_per_gb'] +\
                        grouped_df_ultra['over_messages'] * grouped_df_ultra['rub_per_message']
grouped_df_ultra.head()
```

```python
grouped_df_smart['fee'] = grouped_df_smart['rub_monthly_fee'] +\
                        grouped_df_smart['over_minutes'] * grouped_df_smart['rub_per_minute'] +\
                        grouped_df_smart['over_mb'] / 1000 * grouped_df_smart['rub_per_gb'] +\
                        grouped_df_smart['over_messages'] * grouped_df_smart['rub_per_message']
grouped_df_smart.head()
```

### 2.8.Вывод

Мы сгруппировали нужные нам данные по месяцам и id пользователей (используя метод **groupby**), скомбинировали (с помощью метода **merge**) несколько таблиц и получили одну (**grouped_df**), которая содержала всю нужную нам информацию. Затем ее мы поделили на две таблицы по тарифам - **grouped_df_ultra** и  **grouped_df_smart**. Мы расчитали выручку (сверх абонентской платы по тарифам) для каждого тарифа.

По такому же принципу мы будем группировать данные по израсходованному интернет-трафику и отправленным сообщениям.

### 3.1. Сколько минут разговора требуется пользователям в месяц?

```python
ultra_calls = grouped_df_ultra.pivot_table(index=['user_id','month_x'], values='duration')
smart_calls = grouped_df_smart.pivot_table(index=['user_id','month_x'], values='duration')
```

 ## Шаг 4.Проверка гипотез

### 4.1. Средняя выручка пользователей тарифов «Ультра» и «Смарт» равна

Поскольку мы имеем две генеральных совокупности, нам нужно сопоставить их среднее. 

Нулевая гипотеза мы сформулируем так: **"Средняя выручка пользователей тарифов «Ультра» и «Смарт» равна"**. 

Соответственно, альтернативная гипотеза будет: **"Средняя выручка пользователей тарифов «Ультра» и «Смарт» не равна"**.

Чтобы проверить гипотезу о равенстве среднего двух генеральных совокупностей, применим метод **scipy.stats.ttest_ind**: 

```python
alpha = .05

results = st.ttest_ind(
    grouped_df_ultra['fee'], 
    grouped_df_smart['fee'])

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

По резутатам теста видим,что вероятность того, что средняя выручка пользователей двух тарифов равна, весьма мала. 

### 4.2. Средняя выручка пользователей тарифа "Ультра""Смарт" из Москвы и из других регионов равна

Отфильтруем таблицы **grouped_df_ultra** и **grouped_df_smart** на абонентов из Москвы и из других городов (с помощью метода **query**):

```python
grouped_df_ultra_M = grouped_df_ultra.query('city == "Москва"').reset_index(drop=True)
grouped_df_ultra_M.head()
```

```python
grouped_df_smart_M = grouped_df_smart.query('city == "Москва"').reset_index(drop=True)
grouped_df_smart_M.head()
```

```python
grouped_df_ultra_cities = grouped_df_ultra.query('city != "Москва"').reset_index(drop=True)
grouped_df_ultra_cities.head()
```

```python
grouped_df_smart_cities = grouped_df_smart.query('city != "Москва"').reset_index(drop=True)
grouped_df_smart_cities.head()
```

Нулевую гипотезу формулируем так:  **"Средняя выручка пользователей тарифа "Ультра"\"Смарт" из Москвы и из других регионов равна"**

Альтернативная гипотеза:  **"Средняя выручка пользователей тарифа "Ультра"\"Смарт" из Москвы и из других регионов не равна"**

```python
alpha = .05

results = st.ttest_ind(
    grouped_df_ultra_M['fee'], 
    grouped_df_ultra_cities['fee'])

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

```python
alpha = .05

results = st.ttest_ind(
    grouped_df_smart_M['fee'], 
    grouped_df_smart_cities['fee'])

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

По обоим тарифам вероятность того, что средняя выручка пользователей из Москвы и из других регионов равна, весьма высока - у тарифа "Ультра" она составляет почти 89%.
